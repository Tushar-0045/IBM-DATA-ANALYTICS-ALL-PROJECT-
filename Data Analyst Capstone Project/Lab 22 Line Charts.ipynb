### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
# Group by age and calculate median compensation
age_comp = df.groupby('Age')['ConvertedCompYearly'].median()

# Create line chart
plt.figure(figsize=(10,5))
age_comp.plot(kind='line', marker='o')
plt.title('Median Yearly Compensation by Age Group')
plt.ylabel('Compensation (USD)')
plt.xlabel('Age Group')
plt.grid(True)
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
# Filter for ages 25-45 and remove missing values
age_filter = df[
    (df['Age'].isin(['25-29', '30-34', '35-39', '40-44'])) & 
    (df['ConvertedCompYearly'].notna())
]

# Calculate median compensation by age group
age_comp = age_filter.groupby('Age')['ConvertedCompYearly'].median().sort_index()

# Create and customize the line chart
plt.figure(figsize=(10, 6))
age_comp.plot(
    kind='line',
    marker='o',
    color='#2ca02c',  # Better green color
    linewidth=2.5,
    markersize=8
)

# Add chart elements
plt.title('Median Compensation by Age Group (25-45)', pad=20, fontsize=14)
plt.ylabel('Annual Compensation (USD)', fontsize=12)
plt.xlabel('Age Group', fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, alpha=0.3)

# Format y-axis as currency
plt.gca().yaxis.set_major_formatter('${x:,.0f}')

# Ensure proper layout
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
# Use YearsCodePro as experience metric
exp_satisfaction = df.groupby('YearsCodePro')['JobSatPoints_6'].median()

# Create line chart
plt.figure(figsize=(10,5))
exp_satisfaction.plot(kind='line', marker='s')
plt.title('Job Satisfaction by Years of Professional Coding')
plt.ylabel('Satisfaction Score')
plt.xlabel('Years of Experience')
plt.grid(True)
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
# Group by WorkExp if available, otherwise use YearsCodePro
exp_col = 'WorkExp' if 'WorkExp' in df.columns else 'YearsCodePro'
exp_comp = df.groupby(exp_col)['ConvertedCompYearly'].median()

# Create line chart
plt.figure(figsize=(10,5))
exp_comp.plot(kind='line', marker='D', color='purple')
plt.title('Median Compensation by Experience Level')
plt.ylabel('Compensation (USD)')
plt.xlabel('Experience Level')
plt.grid(True)
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
# Group by experience
exp_satisfaction = df.groupby(exp_col)['JobSatPoints_6'].median()

# Create line chart
plt.figure(figsize=(10,5))
exp_satisfaction.plot(kind='line', marker='^', color='orange')
plt.title('Job Satisfaction by Experience Level')
plt.ylabel('Satisfaction Score')
plt.xlabel('Experience Level')
plt.grid(True)
plt.show()
```
