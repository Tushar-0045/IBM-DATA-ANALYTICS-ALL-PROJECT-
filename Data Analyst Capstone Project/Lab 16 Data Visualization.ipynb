**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

Plot a histogram of CompTotal (Total Compensation).

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert Age to numeric if it's categorical
if df['Age'].dtype == object:
    age_map = {
        'Under 18 years old': 17,
        '18-24 years old': 21,
        '25-34 years old': 30,
        '35-44 years old': 40,
        '45-54 years old': 50,
        '55-64 years old': 60,
        '65 years or older': 70
    }
    df['AgeNumeric'] = df['Age'].map(age_map)
else:
    df['AgeNumeric'] = df['Age']

# Check for compensation columns
comp_col = None
for col in ['ConvertedCompYearly', 'CompTotal', 'Compensation']:
    if col in df.columns:
        comp_col = col
        break

if comp_col:
    # Filter for age range 45-60 and remove outliers
    age_range = df[(df['AgeNumeric'] >= 45) & (df['AgeNumeric'] <= 60)]
    
    # Calculate median compensation by age
    median_comp = age_range.groupby('AgeNumeric')[comp_col].median()
    
    # Create line chart
    plt.figure(figsize=(10, 6))
    median_comp.plot.line(
        marker='o',
        markersize=8,
        linewidth=2,
        color='royalblue'
    )
    
    # Customize the plot
    plt.title(f'Median {comp_col} by Age (45-60)')
    plt.xlabel('Age')
    plt.ylabel('Median Compensation (USD)')
    plt.grid(True, linestyle='--', alpha=0.6)
    
    # Format y-axis as currency
    plt.gca().set_yticklabels(['${:,.0f}'.format(x) for x in plt.gca().get_yticks()])
    
    plt.tight_layout()
    plt.show()
else:
    print("No compensation column found. Available columns:", df.columns.tolist())
```
